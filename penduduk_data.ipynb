AMBIL DATA 2018

```python
import pandas as pd

# 1. Baca data
df = pd.read_csv(
    "Jumlah Penduduk Menurut Jenis Kelamin dan Kabupaten_Kota Provinsi Jawa Timur, 2018.csv"
)

# 2. Rename kolom (karena header-nya berantakan)
df.columns = [
    'nama_kabupaten_kota',
    'laki_laki',
    'perempuan',
    'total_existing'
]

# 3. Buang baris header tambahan
df = df.iloc[3:].reset_index(drop=True)

# 4. Ubah ke numerik
df['laki_laki'] = pd.to_numeric(df['laki_laki'], errors='coerce')
df['perempuan'] = pd.to_numeric(df['perempuan'], errors='coerce')

# 5. Hitung jumlah penduduk
df['jumlah_penduduk'] = df['laki_laki'] + df['perempuan']

# 6. Ambil kolom final
df_penduduk_2018 = df[['nama_kabupaten_kota', 'jumlah_penduduk']]

df_penduduk_2018 = df_penduduk_2018[
    df_penduduk_2018['nama_kabupaten_kota'].str.strip().str.lower() != 'jawa timur'
]

# Tambahkan kolom tahun
df_penduduk_2018['tahun'] = 2018

df_penduduk_2018 = df_penduduk_2018[
    ['nama_kabupaten_kota', 'tahun', 'jumlah_penduduk']
]

df_penduduk_2018
```

AMBIL DATA 2024

```python
import pandas as pd

# 1. BACA DATA
df = pd.read_csv('lj_prtmbhn_dn_kpdtn_pnddk_mnrt_kbptn_kt_cl.csv')

# 2. STANDARISASI
df.columns = df.columns.str.strip().str.lower()

df['kategori'] = df['kategori'].str.lower().str.strip()
df['kabupatenkota'] = df['kabupatenkota'].str.title().str.strip()

df['jumlah'] = pd.to_numeric(df['jumlah'], errors='coerce')
df['tahun'] = df['tahun'].astype(int)

# 3. AMBIL HANYA JUMLAH PENDUDUK
df_penduduk_2024 = df[df['kategori'] == 'jumlah_penduduk']

# 4. PILIH KOLOM YANG DIPERLUKAN
df_penduduk_2024 = df_penduduk_2024[
    ['tahun', 'kabupatenkota', 'jumlah']
].rename(columns={
    'kabupatenkota': 'nama_kabupaten_kota',
    'jumlah': 'jumlah_penduduk'
})

# 5. FILTER TAHUN 2024
df_penduduk_2024 = df_penduduk_2024[
    df_penduduk_2024['tahun'] == 2024
]

# 6. HAPUS DATA PROVINSI (JAWA TIMUR)
df_penduduk_2024 = df_penduduk_2024[
    df_penduduk_2024['nama_kabupaten_kota'] != 'Jawa Timur'
]

# 7. SIAPKAN DATA TAHUN SEBELUMNYA (2023)
df_prev = df_penduduk_2023[
    ['nama_kabupaten_kota', 'jumlah_penduduk']
].rename(columns={
    'jumlah_penduduk': 'jumlah_penduduk_prev'
})

# 8. UBAH RIBU JIWA KE SATUAN JIWA (HANDLING BY 2 ANGKA TERAKHIR TAHUN SEBELUMNYA)
df_penduduk_2024 = df_penduduk_2024.merge(
    df_prev,
    on='nama_kabupaten_kota',
    how='left'
)
df_penduduk_2024['jumlah_penduduk'] = (
    (df_penduduk_2024['jumlah_penduduk'] * 1000)
    .astype(int)
    // 100 * 100
    +
    (df_penduduk_2024['jumlah_penduduk_prev'] % 100)
)

#9. SORT
df_penduduk_2024 = df_penduduk_2024[
    ['nama_kabupaten_kota', 'tahun', 'jumlah_penduduk']
]

df_penduduk_2024
```

ISI KODE KABUPATEN KOTA

```python
import pandas as pd

# =====================================================
# A. DATA MASTER KODE WILAYAH 
# =====================================================

df_master = pd.read_csv('jumlah_tenaga_kesehatan.csv')

# rapikan nama kolom
df_master.columns = df_master.columns.str.strip()

# standarisasi nama kabupaten
df_master['nama_kabupaten_kota'] = (
    df_master['nama_kabupaten_kota']
    .astype(str)
    .str.strip()
    .str.upper()
)

# ambil mapping
master_kabkot = (
    df_master[['kode_kabupaten_kota', 'nama_kabupaten_kota']]
    .drop_duplicates()
)

master_kabkot
```

```python
# =====================================================
# B. SAMAKAN FORMAT NAMA KAB/KOTA
# =====================================================
df_penduduk_all['nama_kabupaten_kota'] = (
    df_penduduk_all['nama_kabupaten_kota']
    .str.strip()
    .str.title()
)

master_kabkot['nama_kabupaten_kota'] = (
    master_kabkot['nama_kabupaten_kota']
    .str.strip()
    .str.title()
)
```

```python
# =====================================================
# C. MERGE KODE KABUPATEN 
# =====================================================

# pastikan data TIDAK punya kolom kode
df_penduduk_all = df_penduduk_all.drop(
    columns=['kode_kabupaten_kota'],
    errors='ignore'
)

# merge
df_penduduk_all = df_penduduk_all.merge(
    master_kabkot,
    on='nama_kabupaten_kota',
    how='left'
)

# =====================================================
# D. FORMAT AKHIR
# =====================================================

df_penduduk_all = df_penduduk_all.sort_values(
    ['tahun', 'kode_kabupaten_kota']
).reset_index(drop=True)

# kembalikan format penulisan
df_penduduk_all['nama_kabupaten_kota'] = (
    df_penduduk_all['nama_kabupaten_kota']
    .str.title()
)
df_penduduk_all = df_penduduk_all[
    [
        'kode_kabupaten_kota',
        'nama_kabupaten_kota',
        'tahun',
        'jumlah_penduduk'
    ]
]

df_penduduk_all
```

SIMPAN DATA CSV

```python
df_penduduk_all.to_csv('data_penduduk.csv', index=False)
```
