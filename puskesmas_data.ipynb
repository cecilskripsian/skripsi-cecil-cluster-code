PIVOT DATA RAWAT INAP DAN NON RAWAT INAP

```python
import pandas as pd

# 1. Baca dataset
df = pd.read_csv('jmlh_psksms_brdsrkn_plynn_rwt_np_dn_nn_rwt_np.csv')

# 2. Pastikan 'tahun' adalah string
df['tahun'] = df['tahun'].astype(str)

# 3. Gabung seluruh data per kategori per kabupaten-kota per tahun
df_tahun = (
    df.groupby([
        'kode_provinsi',
        'nama_provinsi',
        'kode_kabupaten_kota',
        'nama_kabupaten_kota',
        'kategori',
        'tahun'
    ])['jumlah']
    .sum()
    .reset_index()
)

# 4. Pivot kolom kategori
df_pivot = df_tahun.pivot_table(
    index=[
        'kode_provinsi',
        'nama_provinsi',
        'kode_kabupaten_kota',
        'nama_kabupaten_kota',
        'tahun'
    ],
    columns='kategori',
    values='jumlah',
    aggfunc='sum'
).reset_index()

# 5. Rename kolom
df_pivot = df_pivot.rename(columns={
    'PELAYANAN RAWAT INAP': 'PELAYANAN_RAWAT_INAP',
    'PELAYANAN NON RAWAT INAP': 'PELAYANAN_NON_RAWAT_INAP'
})

# 6. Urutkan
df_pivot = df_pivot.sort_values(
    ['tahun', 'kode_kabupaten_kota'],
    ascending=[True, True]
).reset_index(drop=True)

# 7. Tampilkan
df_pivot
```
