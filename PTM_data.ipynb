SUM JUMLAH PENYAKIT TIDAK MENULAR

```python
import pandas as pd

# 1. Baca dataset
df = pd.read_csv('jumlah_penyakit_tidak_menular.csv')

# 2. cek periode_update STRING
df['periode_update'] = df['periode_update'].astype(str)

# 3. Ambil tahun
df['tahun'] = df['periode_update'].str[:4]

# 4. Jumlahkan seluruh kategori penyakit per tahun per kabupaten
df_tahun_kab = (
    df.groupby([
        'kode_kabupaten_kota',
        'nama_kabupaten_kota',
        'tahun'
    ])['jumlah']
    .sum()
    .reset_index()
)
df_tahun_kab = df_tahun_kab.rename(columns={
    'jumlah': 'JUMLAH_PENYAKIT_TIDAK_MENULAR'
})
# 5. Urutkan berdasarkan tahun 
df_tahun_kab = df_tahun_kab.sort_values(
    ['tahun', 'kode_kabupaten_kota'],
    ascending=[True, True]
).reset_index(drop=True)

df_tahun_kab['JUMLAH_PENYAKIT_TIDAK_MENULAR'] = (
    df_tahun_kab['JUMLAH_PENYAKIT_TIDAK_MENULAR']
    .round()
    .astype('Int64')
)
df_tahun_kab.head()
```
