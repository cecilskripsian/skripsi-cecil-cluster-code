AMBIL DATA 2018-2022

```python
import pandas as pd

# 1. Baca dataset
df = pd.read_csv('jumlah_tenaga_kesehatan.csv')

# 2. cek periode_update STRING
df['periode_update'] = df['periode_update'].astype(str)

# 3. Ambil tahun
df['tahun'] = df['periode_update'].str[:4]

# 4. Jumlahkan seluruh kategori penyakit per tahun per kabupaten
df_tahun_kab_2018 = (
    df.groupby([
        'kode_kabupaten_kota',
        'nama_kabupaten_kota',
        'tahun'
    ])['jumlah']
    .sum()
    .reset_index()
)
df_tahun_kab_2018 = df_tahun_kab_2018.rename(columns={
    'jumlah': 'JUMLAH_TENAGA_KESEHATAN'
})

# 5. Urutkan berdasarkan tahun ASC dan nama kabupaten-kota ASC
df_tahun_kab_2018 = df_tahun_kab_2018.sort_values(
    ['tahun', 'kode_kabupaten_kota'],
    ascending=[True, True]
).reset_index(drop=True)

df_tahun_kab_2018
```

AMBIL DATA 2023-2024

```python
import pandas as pd

df = pd.read_csv('jmlh_tng_kshtn_mnrt_kbptn_kt_cl.csv')

# 1. Tahun string
df['tahun'] = df['tahun'].astype(str)

# 2. Ambil tahun 2023 & 2024
df = df[df['tahun'].isin(['2023', '2024'])]

# 3. Hapus baris provinsi
df = df[
    ~df['kabupatenkota']
      .str.strip()
      .str.upper()
      .eq('JAWA TIMUR')
]

# 4. Jumlahkan per kabupaten per tahun
df_tahun_kab_2023 = (
    df.groupby(
        ['kabupatenkota', 'tahun']
    )['jumlah']
    .sum()
    .reset_index()
)

# 5. Rename kolom
df_tahun_kab_2023 = df_tahun_kab_2023.rename(columns={
    'kabupatenkota': 'nama_kabupaten_kota',
    'jumlah': 'JUMLAH_TENAGA_KESEHATAN'
})

# 6. Sort
df_tahun_kab_2023 = df_tahun_kab_2023.sort_values(
    ['tahun', 'nama_kabupaten_kota'],
    ascending=[True, True]
).reset_index(drop=True)

df_tahun_kab_2023['JUMLAH_TENAGA_KESEHATAN'] = (
    df_tahun_kab_2023['JUMLAH_TENAGA_KESEHATAN']
    .round()
    .astype('Int64')
)

# 7. Tampilkan hasil
df_tahun_kab_2023
```

ISI KODE KABUPATEN KOTA

```python
import pandas as pd

# =====================================================
# DATA MASTER KODE WILAYAH 
# =====================================================

df_master = pd.read_csv('jumlah_tenaga_kesehatan.csv')

# rapikan nama kolom
df_master.columns = df_master.columns.str.strip()

# standarisasi nama kabupaten
df_master['nama_kabupaten_kota'] = (
    df_master['nama_kabupaten_kota']
    .astype(str)
    .str.strip()
    .str.upper()
)

# ambil mapping
master_kabkot = (
    df_master[['kode_kabupaten_kota', 'nama_kabupaten_kota']]
    .drop_duplicates()
)

master_kabkot
```

```python
# =====================================================
# DATA TENAGA KESEHATAN (2023–2024)
# =====================================================

df = pd.read_csv('jmlh_tng_kshtn_mnrt_kbptn_kt_cl.csv')

# rapikan nama kolom
df.columns = df.columns.str.strip()

# tahun string
df['tahun'] = df['tahun'].astype(str)

# filter tahun
df = df[df['tahun'].isin(['2023', '2024'])]

# hapus agregat provinsi
df = df[
    ~df['kabupatenkota']
      .astype(str)
      .str.strip()
      .str.upper()
      .eq('JAWA TIMUR')
]

# standarisasi nama kabupaten
df['nama_kabupaten_kota'] = (
    df['kabupatenkota']
    .astype(str)
    .str.strip()
    .str.upper()
)

# =====================================================
# AGREGASI PER KABUPATEN PER TAHUN
# =====================================================

df_tahun_kab_2023 = (
    df.groupby(
        ['nama_kabupaten_kota', 'tahun']
    )['jumlah']
    .sum()
    .reset_index()
)

df_tahun_kab_2023 = df_tahun_kab_2023.rename(columns={
    'jumlah': 'JUMLAH_TENAGA_KESEHATAN'
})

# =====================================================
# MERGE KODE KABUPATEN 
# =====================================================

# pastikan data TIDAK punya kolom kode
df_tahun_kab_2023 = df_tahun_kab_2023.drop(
    columns=['kode_kabupaten_kota'],
    errors='ignore'
)

# merge
df_tahun_kab_2023 = df_tahun_kab_2023.merge(
    master_kabkot,
    on='nama_kabupaten_kota',
    how='left'
)

# =====================================================
# FORMAT AKHIR
# =====================================================

df_tahun_kab_2023 = df_tahun_kab_2023.sort_values(
    ['tahun', 'kode_kabupaten_kota']
).reset_index(drop=True)

# kembalikan format penulisan
df_tahun_kab_2023['nama_kabupaten_kota'] = (
    df_tahun_kab_2023['nama_kabupaten_kota']
    .str.title()
)
df_tahun_kab_2023 = df_tahun_kab_2023[
    [
        'kode_kabupaten_kota',
        'nama_kabupaten_kota',
        'tahun',
        'JUMLAH_TENAGA_KESEHATAN'
    ]
]

df_tahun_kab_2023['JUMLAH_TENAGA_KESEHATAN'] = (
    df_tahun_kab_2023['JUMLAH_TENAGA_KESEHATAN']
    .round()
    .astype('Int64')  )

df_tahun_kab_2023
```

```python
# =====================================================
# GABUNG DATA 
# =====================================================

df_nakes = pd.concat(
    [
        df_tahun_kab_2018,
        df_tahun_kab_2023
    ],
    axis=0,
    ignore_index=True
)

# urutkan
df_nakes = df_nakes.sort_values(
    ['tahun', 'kode_kabupaten_kota']
).reset_index(drop=True)

df_nakes
```
